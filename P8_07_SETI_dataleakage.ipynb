```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train_labels = pd.read_csv("/content/drive/MyDrive/Gaetan_Travail/ML/Formation_OC_IML/P8/train_labels.csv")
submission = pd.read_csv("/content/drive/MyDrive/Gaetan_Travail/ML/Formation_OC_IML/P8/sample_submission.csv")

leak_train = pd.read_csv("/content/drive/MyDrive/Gaetan_Travail/ML/Formation_OC_IML/P8/leaks/leak_train.csv")
leak_test = pd.read_csv("/content/drive/MyDrive/Gaetan_Travail/ML/Formation_OC_IML/P8/leaks/leak_test.csv")
```

```python
print(leak_train.shape)
print(train_labels.shape)

print(leak_test.shape)
print(submission.shape)
```

```python
pd.merge(leak_train, train_labels, how="left", on="id").head()
```

```python
tmp = pd.merge(leak_train, train_labels, how="left", on="id")

x_train, x_valid, y_train, y_valid = train_test_split(tmp.iloc[:, :-2], tmp.target, test_size=0.15, stratify=tmp.target, random_state=0)

print(x_train.shape)
print(x_valid.shape)
print(y_train.shape)
print(y_valid.shape)
```

```python
# AutoGluon
```

```python
data = TabularDataset(pd.concat([x_train, y_train], axis=1)).reset_index(drop=True)
```

```python
predictor = TabularPredictor(
    label="target",
    problem_type="binary",
    eval_metric="roc_auc",
    sample_weight="auto_weight"
).fit(
   data,
   time_limit=600,
   presets='best_quality'
)
```

```python
performance = predictor.evaluate(TabularDataset(pd.concat([x_valid, y_valid], axis=1)).reset_index(drop=True))
```

```python
predictor.leaderboard(TabularDataset(pd.concat([x_valid, y_valid], axis=1)).reset_index(drop=True), silent=True)
```
